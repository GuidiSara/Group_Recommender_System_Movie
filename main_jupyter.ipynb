### Neighbours

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

## Find neighbors in common from two person

### Candidate a group

```python
def candidate_group_person():
    people_tot = []
    found = True

    while found:
        person_one = random.randint(1, 943)
        person_two = random.randint(1, 943)
        if(person_one != person_two and (ratings['user_id'] == person_one).any() and (ratings['user_id'] == person_two).any()):
            found = False
            people_tot.append(person_one)
            people_tot.append(person_two)

    return people_tot
```

```python
def find_common_elements(series1, series2):
    common_elements = series1[series1.isin(series2)]
    return common_elements
```

```python
people=[0,1]
sim_user_30_u1 = sim_user_30_u.iloc[0]
sim_user_30_u2 = sim_user_30_u.iloc[1]
find_neighbors = find_common_elements(sim_user_30_u1, sim_user_30_u2)
while find_neighbors.empty :
    people = candidate_group_person()
    sim_user_30_u1 = sim_user_30_u.iloc[people[0]]
    sim_user_30_u2 = sim_user_30_u.iloc[people[1]]
    find_neighbors = find_common_elements(sim_user_30_u1, sim_user_30_u2)
    
```

```python
neighbors_dataFrame = pd.DataFrame(columns=['i_u1', 'i_u2', 'value', 'weighted_average'])

for index, value in find_neighbors.iteritems():
    i_u1 = sim_user_30_u1.index[sim_user_30_u1 == value][0]
    i_u1 = i_u1.split("top")
    i_u1 = i_u1[1]
    i_u2 = sim_user_30_u2.index[sim_user_30_u2 == value][0]
    i_u2 = i_u2.split("top")
    i_u2 = i_u2[1]
    weighted_average = int(i_u1) + int(i_u2)
    new_rows = pd.DataFrame([[ i_u1, i_u2, value, weighted_average]], columns=['i_u1', 'i_u2', 'value', 'weighted_average'])
    neighbors_dataFrame = pd.concat([neighbors_dataFrame, new_rows], ignore_index=True)
    find_neighbors = find_neighbors.drop(index=index)

neighbors_dataFrame['weighted_average'] = neighbors_dataFrame['weighted_average'].astype(float)    
neighbors_dataFrame
```
